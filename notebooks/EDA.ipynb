```python
from datetime import datetime, timedelta
# for Time_Orderd null Value we can use Time_Order_picked value subtracting 10 min because we analysis that Time_Orderd value less than Time_Order_picked value
# we analysis that Time_Orderd most of value  less than Time_Order_picked value are 5 min, 10 min and 15 min we so we used 10 min
def gettime_from_Time_Order_picked(time_str):
    try:
        time_obj = datetime.strptime(time_str, "%H:%M") # convert string to datetime object
        new_time_obj = time_obj - timedelta(minutes=10) # subtract 15 minutes from datetime object
        new_time_str = new_time_obj.strftime("%H:%M") # convert datetime object back to string
        return new_time_str
    except Exception as e:
        return new_time_str
# for Time_Order_picked null Value we can use Time_Orderd value Adding 10 min because we analysis that Time_Order_picked value greater than Time_Orderd  value
# we analysis that Time_Order_picked most of value  Greater than Time_Orderd value are 5 min, 10 min and 15 min we so we used 10 min
def gettime_from_Time_Order(time_str):
    try:
        time_obj = datetime.strptime(time_str, "%H:%M") # convert string to datetime object
        new_time_obj = time_obj + timedelta(minutes=10) # add 15 minutes from datetime object
        new_time_str = new_time_obj.strftime("%H:%M") # convert datetime object back to string
        return new_time_str
    except Exception as e:
        return time_str

```

```python
def convert_time_24_to_00(time_str):
    split_val = time_str.split(":")
    if(str(split_val[0]) == "24"):
        new_val = "00:"+ str(split_val[1])
        return new_val
    else:
        return time_str
```

```python
def get_fillna_value_for_Road_traffic_density(value):
    #after analyzing we get the below type of trafic density respet to the time
    high_time_range = [11,12,13,14,15]
    mediam_time_range = [16,17,18,19]
    jam_time_range = [19,20,21,22]
    trrffic_condition = ""
    if(value in high_time_range):
        trrffic_condition = "High"
    elif(value in jam_time_range):
        trrffic_condition = "Jam"
    elif(value in mediam_time_range):
        trrffic_condition = "Medium"
    else:
        trrffic_condition = "Low"
    return trrffic_condition
```

```python
import datetime
def get_weekday_from_date(date_str):

    # date_str = '2023-04-26' # date in yyyy-mm-dd format
    date = datetime.datetime.strptime(date_str, '%d-%m-%Y')
    week_day = date.strftime('%A') # %A returns the full weekday name

    return week_day
```

```python
#convert Order Time to Number 
import datetime
def convert_time_to_number(time_str):
    time_str_dt = datetime.datetime.strptime(time_str, '%H:%M') # convert to datetime object
    time_str_seconds = (time_str_dt - time_str_dt.replace(hour=0, minute=0, second=0, microsecond=0)).total_seconds() # calculate total seconds since midnight
    return time_str_seconds
```
